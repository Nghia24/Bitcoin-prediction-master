```python
import numpy as np 
import pandas as pd
from datetime import datetime
from datetime import timedelta
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('export_dataframe1.csv')
data['Date'] = pd.to_datetime(data['time'],unit='s').dt.date
data = data.set_index('Date')
data.head(10)
```

```python
df = data.iloc[:, 7:8].copy()
df.head(10)
```

```python
n = 10 
t = 0.8 
split = int (t*len(df))
```

```python
#A variable for predicting 'n' days out into the future
prediction_days = 5
#Create another column (the target or dependent variable) shifted 'n' units up
df['Prediction'] = df[['weighted_price']].shift(-prediction_days)
```

```python
#CREATE THE INDEPENDENT DATA SET (X)

# Convert the dataframe to a numpy array and drop the prediction column
X = np.array(df.drop(['Prediction'],1))

#Remove the last 'n' rows where 'n' is the prediction_days
X= X[:len(df)-prediction_days]
print(X)
```

```python
#CREATE THE DEPENDENT DATA SET (y) 
# Convert the dataframe to a numpy array (All of the values including the NaN's) 
y = np.array(df['Prediction'])  
# Get all of the y values except the last 'n' rows 
y = y[:-prediction_days] 
print(y)
```

```python
# Split the data into 80% training and 20% testing
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state = 0,shuffle=False)
```

```python
print(y_test)
```

```python
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
r2 = r2_score(x_test, y_test)
mse = mean_squared_error(x_test, y_test)
mae = mean_absolute_error(x_test, y_test)
print('\n')
print('Printing Mean Squared Error of Predictions...')
print('Test MSE: %.6f' % mse)
print('Test R2: %.6f' % r2)
print('Test MAE: %.6f' % mae)
```

```python
# Set prediction_days_array equal to the last 30 rows of the original data set from the price column
prediction_days_array = np.array(df.drop(['Prediction'],1))[-prediction_days:]
print(prediction_days_array)
```

```python
from sklearn.svm import SVR
# Create and train the Support Vector Machine 
svr_rbf = SVR(kernel='rbf', C=1e3, gamma=0.00001)#Create the model
svr_rbf.fit(x_train, y_train) #Train the model
```

```python
# Testing Model: Score returns the accuracy of the prediction. 
# The best possible score is 1.0
svr_rbf_confidence = svr_rbf.score(x_test, y_test)
print("svr_rbf accuracy: ", svr_rbf_confidence)
```

```python
# Print the predicted value
svm_prediction = svr_rbf.predict(x_test)
print(svm_prediction)

print()

#Print the actual values
print(y_test)
```

```python
# Print the model predictions for the next 'n=30' days
svm_prediction = svr_rbf.predict(prediction_days_array)
print(svm_prediction)
```

```python
import matplotlib.pyplot as plt
def line_plot(line1, line2, title=''):
    fig, ax = plt.subplots(1, figsize=(20, 10))
    ax.plot(line1, 'b', label='label1', linewidth=3)
    ax.plot(line2, 'g', label='label2', linewidth=2)
    ax.set_ylabel('price [USD]', fontsize=14)
    ax.set_title(title, fontsize=16)
line_plot(x_test, y_test)
```
